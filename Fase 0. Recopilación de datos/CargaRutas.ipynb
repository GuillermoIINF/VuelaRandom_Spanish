# Fase 0: <i>Recopilación de datos</i>
## <b>Obtención de las rutas aéreas existentes</b>

&emsp;&emsp;Se procede a obtener la información de los aeropuertos (<i>ciudad</i>, <i>país</i>, <i>código IATA</i> y <i>coordenadas</i> del aeropuerto) y las rutas aéreas sobre las que vamos a realizar el webscraping

```python
import pandas as pd
import numpy as np
rutasDat = "../../data/routes.dat"
rutasCsv = "../../data/rutasIniciales.csv"
aeropuertosDat = "../../data/airports.dat"
aeropuertosCsv = "../../data/aeropuertos_data.csv"
```

&emsp;&emsp;En primer lugar, obtenemos las rutas aéreas con origen en <b><i>Barcelona</i></b> y <b><i>Madrid</i></b> y las guardamos en un fichero CSV para poder tenerlos disponibles a la hora de buscar vuelos.

```python
dfRutas = pd.read_csv(rutasDat, names=["Airline", "Airline ID", "Source Airport", "Source Airport ID", "Destination Airport", "Destination Airport ID", "Codeshare", "Stops", "Equipment"])
dfRutasIniciales=dfRutas[dfRutas["Source Airport"].isin(["MAD", "BCN"])]
dfRutasIniciales.to_csv(rutasCsv)
print("rutasIniciales.csv creado correctamente")
dfRutasIniciales
```

&emsp;&emsp;En segundo lugar, obtenemos los datos de los aeropuertos.

```python
dfAeropuertos = pd.read_csv(aeropuertosDat, header=None)
dfAeropuertos = dfAeropuertos.replace('\\N', np.nan)
dfAeropuertos.dropna(axis=0, inplace = True)
dfAeropuertosSelecc = dfAeropuertos[[2,3,4,6,7]]
dfAeropuertosSelecc.columns=["CIUDAD","PAIS","COD","LATITUD","LONGITUD"]
dfAeropuertosSelecc.to_csv(aeropuertosCsv)
print("aeropuertos_data.csv creado correctamente")
dfAeropuertosSelecc
```
